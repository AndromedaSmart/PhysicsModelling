```python
import matplotlib.pyplot as plt
import numpy as np
import math

def function1(i, width, x):
    return np.sqrt(2 / width) * np.sin(i * np.pi * x / width)

def function2(h, i, massOfElectron, width):
    return ((h ** 2 * math.pi ** 2 * i**2) / (2 * massOfElectron * width ** 2))

if __name__ == '__main__':
    n = 5

    h_p = 6.626e-34
    massOfElectron = 9.11e-31
    width = 5**-9
    x_list = np.linspace(0, width, 100)
    s = 10
    plt.figure(figsize=(15, 10))

    for i in range(1, n + 1):
        psi_list = list(map(lambda x: function1(i, width, x), x_list))
        En = function2(h_p, i, massOfElectron, width)

        plt.subplot(3, 3, 2 * i - 1)
        plt.plot(x_list, psi_list, "-")
        plt.xlabel('l', fontsize=13)
        plt.ylabel('', fontsize=13)
        plt.xticks(np.arange(0, width, step=0.1))
        print("Energy = " + str(En) + ", if n = " + str(i))
        plt.grid()

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
